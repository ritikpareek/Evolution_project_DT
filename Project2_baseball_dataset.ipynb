# Removing Outliers

```python
from scipy.stats import zscore
outliers=np.abs(zscore(df))
print(np.where(outliers>3))
thresold=3
```

```python
df_new=df[(outliers<3).all(axis=1)]
```

```python
df.shape
```

```python
df_new.shape
```

1 row is removed as it includes outliers in it

# Skewness

```python
df_new.skew()
```

For a good model the skewness should be bellow 0.55...but in this data some variables are positively skeweed we have to make them normal by using log transformation

```python
for col in df_new.columns:
    if df_new.skew().loc[col]>0.55:
        df_new[col]=np.log1p(df_new[col])
```

```python
df_new.skew()
```

now all the variables are normally distributed

# Regression Process

In this dataset the target variable is Continuous variable so that the regression process should be implimented

```python
from sklearn.model_selection import train_test_split,cross_val_score,GridSearchCV
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
```

```python
def max_r2_score(rgr,x,y):
    max_score=0
    for r in range(42,101):
        
        x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.20,random_state=r)
        rgr.fit(x_train,y_train)
        pred=rgr.predict(x_test)
        r_sc=r2_score(y_test,pred)
        if r_sc>max_score:
            max_score=r_sc
            r_state=r
            
    print('max score at random_state of',r_state,'is',max_score)
    return r_state
```

```python
lr=LinearRegression()
sc=max_r2_score(lr,x,y)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.20,random_state=73)
lr.fit(x_train,y_train)
pred=lr.predict(x_test)
```

```python
pred_y=lr.predict(x_train)
r2_score(y_train,pred_y)
```

```python
print(r2_score(y_test,pred))
print('\n')
print('RMSE',np.sqrt(mean_squared_error(y_test,pred)))
```

```python
score=cross_val_score(lr,x,y,cv=5,scoring='r2')
print(score)
print(score.mean())
print(score.std())
```

# Lasso Regression

```python
from sklearn.linear_model import Lasso
lsr=Lasso()
pr={'alpha':[.0001,.001,.01,.1,1]}
gs=GridSearchCV(lsr,pr,cv=5)
gs.fit(x,y)
gs.best_params_
```

```python
lsr=Lasso(alpha=1)
sc=max_r2_score(lsr,x,y)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.20,random_state=82)
lsr.fit(x_train,y_train)
pred=lsr.predict(x_test)
```

```python
print(r2_score(y_test,pred))
print('\n')
print('RMSE',np.sqrt(mean_squared_error(y_test,pred)))
```

```python
score=cross_val_score(lsr,x,y,cv=5,scoring='r2')
print(score)
print(score.mean())
print(score.std())
```
