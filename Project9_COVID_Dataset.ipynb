# Skewness

```python
df_new.skew()
```

```python
for col in df_new.columns:
    if df_new.skew().loc[col]>0.55:
        df_new[col]=np.log1p(df_new[col])
```

```python
df_new.skew()
```

# Setting x and y variable

```python
x=df_new.drop('Deaths',axis=1)
x.shape
```

```python
y=df_new['Deaths']
y.shape
```

# Standardscaler

```python
from sklearn.preprocessing import StandardScaler
scale=StandardScaler()
x=scale.fit_transform(x)
```

```python
x.shape
```

# Regression Process

```python
from sklearn.model_selection import train_test_split,cross_val_score,GridSearchCV
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
```

```python
def max_r2_score(rgr,x,y):
    max_score=0
    for r in range(42,101):
        
        x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.20,random_state=r)
        rgr.fit(x_train,y_train)
        pred=rgr.predict(x_test)
        r_sc=r2_score(y_test,pred)
        if r_sc>max_score:
            max_score=r_sc
            r_state=r
            
    print('max score at random_state of',r_state,'is',max_score)
    return r_state
```

```python
lr=LinearRegression()
sc=max_r2_score(lr,x,y)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.20,random_state=64)
lr.fit(x_train,y_train)
pred=lr.predict(x_test)
```

```python
print(r2_score(y_test,pred))
print('\n')
print('RMSE',np.sqrt(mean_squared_error(y_test,pred)))
```

```python
score=cross_val_score(lr,x,y,cv=5,scoring='r2')
print(score)
print(score.mean())
print(score.std())
```

```python
lr_score=r2_score(y_test,pred),cross_val_score(lr,x,y,cv=5).mean()
```

# Lasso

```python
from sklearn.linear_model import Lasso
lsr=Lasso()
pr={'alpha':[.0001,.001,.01,.1,1]}
gs=GridSearchCV(lsr,pr,cv=5)
gs.fit(x,y)
gs.best_params_
```

```python
lsr=Lasso(alpha=.001)
sc=max_r2_score(lsr,x,y)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.20,random_state=64)
lsr.fit(x_train,y_train)
pred=lsr.predict(x_test)
```

```python
print(r2_score(y_test,pred))
print('\n')
print('RMSE',np.sqrt(mean_squared_error(y_test,pred)))
```

```python
score=cross_val_score(lsr,x,y,cv=5,scoring='r2')
print(score)
print(score.mean())
print(score.std())
```

```python
lsr_score=r2_score(y_test,pred),cross_val_score(lsr,x,y,cv=5).mean()
```

# KneighborsRegressor

```python
from sklearn.neighbors import KNeighborsRegressor
knn=KNeighborsRegressor()
pr={'n_neighbors':range(1,10)}
gs=GridSearchCV(knn,pr,cv=5)
gs.fit(x,y)
gs.best_params_
```

```python
#taking odd no as n_neighbors
knn=KNeighborsRegressor(n_neighbors=3)
sc=max_r2_score(knn,x,y)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.20,random_state=49)
knn.fit(x_train,y_train)
pred=knn.predict(x_test)
```

```python
print(r2_score(y_test,pred))
print('\n')
print('RMSE',np.sqrt(mean_squared_error(y_test,pred)))
```

```python
score=cross_val_score(knn,x,y,cv=5,scoring='r2')
print(score)
print(score.mean())
print(score.std())
```

```python
knn_score=r2_score(y_test,pred),cross_val_score(knn,x,y,cv=5).mean()
```

# AdaBoostRegressor

```python
from sklearn.ensemble import AdaBoostRegressor
ada=AdaBoostRegressor()
pr={'learning_rate':[.001,.01,.1,1],'n_estimators':[10,100,500,1000],'base_estimator':[lr,lsr,knn]}
gs=GridSearchCV(ada,pr,cv=5)
gs.fit(x,y)
gs.best_params_
```

```python
ada=AdaBoostRegressor(base_estimator=lsr,learning_rate=1,n_estimators=10)
sc=max_r2_score(ada,x,y)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.20,random_state=64)
ada.fit(x_train,y_train)
pred=ada.predict(x_test)
```

```python
print(r2_score(y_test,pred))
print('\n')
print('RMSE',np.sqrt(mean_squared_error(y_test,pred)))
```

```python
score=cross_val_score(ada,x,y,cv=5,scoring='r2')
print(score)
print(score.mean())
print(score.std())
```

```python
ada_score=r2_score(y_test,pred),cross_val_score(ada,x,y,cv=5).mean()
```
