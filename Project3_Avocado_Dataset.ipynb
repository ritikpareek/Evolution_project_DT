# making Regression model by taking y variable as Averageprice

```python
df1=df[['AveragePrice','Total Volume','4046','4225','4770','Total Bags','Small Bags','Large Bags','XLarge Bags','year','type','region']]
```

```python
#labeling the data
from sklearn.preprocessing import LabelEncoder
scale=LabelEncoder()
df.year=scale.fit_transform(df.year)
df.type=scale.fit_transform(df.type)
df.region=scale.fit_transform(df.region)
```

```python
#removing outliers
from scipy.stats import zscore
z=np.abs(zscore(df1))
df1_new=df1[(z<3).all(axis=1)]
```

```python
df1.shape
```

```python
df1.shape
```

```python
#Removing Skewness
df1.skew()
```

```python
for col in df1.columns:
    if df1.skew().loc[col]>0.55:
        df1[col]=np.log1p(df1[col])
```

```python
df1.skew()
```

```python
#setting X variable
x=df1.iloc[:,1:]
x.shape
```

```python
#Setting Y variable
y=df1.iloc[:,0]
y.shape
```

```python
#Scaling the X variable
from sklearn.preprocessing import StandardScaler
scale=StandardScaler()
x=scale.fit_transform(x)
```

```python
x.shape
```

```python
#importing Liabraries
from sklearn.model_selection import train_test_split,cross_val_score,GridSearchCV
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
```

```python
def max_r2_score(rgr,x,y):
    max_score=0
    for r in range(42,101):
        
        x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.10,random_state=r)
        rgr.fit(x_train,y_train)
        pred=rgr.predict(x_test)
        r_sc=r2_score(y_test,pred)
        if r_sc>max_score:
            max_score=r_sc
            r_state=r
            
    print('max score at random_state of',r_state,'is',max_score)
    return r_state
```

```python
lr=LinearRegression()
sc=max_r2_score(lr,x,y)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.10,random_state=64)
lr.fit(x_train,y_train)
pred=lr.predict(x_test)
```

```python
pred_y=lr.predict(x_train)
r2_score(y_train,pred_y)
```

```python
print(r2_score(y_test,pred))
print('\n')
print('RMSE',np.sqrt(mean_squared_error(y_test,pred)))
```

```python
score=cross_val_score(lr,x,y,cv=5,scoring='r2')
print(score)
print(score.mean())
print(score.std())
```

```python
##Lasso MOdel
from sklearn.linear_model import Lasso
lsr=Lasso()
pr={'alpha':[.0001,.001,.01,.1,1]}
gs=GridSearchCV(lsr,pr,cv=5)
gs.fit(x,y)
gs.best_params_
```

```python
lsr=Lasso(alpha=0.01)
sc=max_r2_score(lsr,x,y)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.10,random_state=83)
lsr.fit(x_train,y_train)
pred=lsr.predict(x_test)
```

```python
print(r2_score(y_test,pred))
print('\n')
print('RMSE',np.sqrt(mean_squared_error(y_test,pred)))
```

```python
score=cross_val_score(lsr,x,y,cv=5,scoring='r2')
print(score)
print(score.mean())
print(score.std())
```

```python
x=df1.iloc[:,0:-1]
x.shape
```

```python
from sklearn.model_selection import train_test_split,cross_val_score,GridSearchCV
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix,roc_curve,roc_auc_score
```

```python
def max_acc_score(clf,x,y):
    max_score=0
    for r in range(42,101):
        
        x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.20,random_state=r)
        clf.fit(x_train,y_train)
        pred=clf.predict(x_test)
        a_sc=accuracy_score(y_test,pred)
        if a_sc>max_score:
            max_score=a_sc
            a_state=r
            
    print('max score at random_state of',a_state,'is',max_score)
    return a_state
```

# logistic Regression

```python
from sklearn.linear_model import LogisticRegression
lg=LogisticRegression()
max_acc_score(lg,x,y)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.20,random_state=48)
lg.fit(x_train,y_train)
pred=lg.predict(x_test)
```

```python
print(accuracy_score(y_test,pred))
print('\n')
print(classification_report(y_test,pred))
print('\n')
print(confusion_matrix(y_test,pred))
```

```python
score=cross_val_score(lg,x,y,cv=5,scoring='accuracy')
print(score)
print(score.mean())
print(score.std())
```
