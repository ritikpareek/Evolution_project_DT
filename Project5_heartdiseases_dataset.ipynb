# Removing Outliers

```python
from scipy.stats import zscore
outliers=np.abs(zscore(df))
print(np.where(outliers>3))
thresold=3
```

```python
df_new=df[(outliers<3).all(axis=1)]
```

```python
df.shape
```

```python
df_new.shape
```

# Skewness

```python
df_new.skew()
```

For a good model the skewness should be bellow 0.55...but in this data some variables are positively skeweed we have to make them normal by using log transformation

```python
for col in df_new.columns:
    if df_new.skew().loc[col]>0.55:
        df_new[col]=np.log1p(df_new[col])
```

```python
df_new.skew()
```

# Setting X and Y variable

```python
x=df_new.iloc[:,0:-1]
x.shape
```

```python
y=df_new.iloc[:,-1]
```

# Classification Process

```python
from sklearn.model_selection import train_test_split,cross_val_score,GridSearchCV
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix,roc_curve,roc_auc_score
```

```python
def max_acc_score(clf,x,y):
    max_score=0
    for r in range(42,101):
        
        x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.10,random_state=r)
        clf.fit(x_train,y_train)
        pred=clf.predict(x_test)
        a_sc=accuracy_score(y_test,pred)
        if a_sc>max_score:
            max_score=a_sc
            a_state=r
            
    print('max score at random_state of',a_state,'is',max_score)
    return a_state
```

```python
from sklearn.linear_model import LogisticRegression
lg=LogisticRegression()
max_acc_score(lg,x,y)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.10,random_state=60)
lg.fit(x_train,y_train)
pred=lg.predict(x_test)
```

```python
print(accuracy_score(y_test,pred))
print('\n')
print(classification_report(y_test,pred))
print('\n')
print(confusion_matrix(y_test,pred))
```

```python
score=cross_val_score(lg,x,y,cv=5,scoring='accuracy')
print(score)
print(score.mean())
print(score.std())
```
